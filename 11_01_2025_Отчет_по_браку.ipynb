##Функция по созданию сводной таблицы на основе столбца и типа продукции

```python
#Функция по созданию сводной таблицы на основе столбца и типа продукции
def pivot_by_columns(release,incongruity,column,type_product):
  #Объединяем выпуск в датафрейм по выбранному столбцу.Установлен фильтр по типу продукции
  if type_product == 0:
    release_by_product_group = release.groupby([column]).agg({"Выпуск,м2":"sum"})
  else:
    release_by_product_group = release[release['Тип_продукции'] == type_product].groupby([column]).agg({"Выпуск,м2":"sum"})
  #Объединяем брак в датафрейм по выбранному столбцу
  if type_product == 0:
    incongruity_by_product_group = incongruity.groupby([column]).agg({"Брак,м2":"sum"})
  else:
    incongruity_by_product_group = incongruity[incongruity['Тип_продукции'] == type_product].groupby([column]).agg({"Брак,м2":"sum"})
  #Используем левое соединение для объединения выпуска и брака
  merged_df = release_by_product_group.merge(incongruity_by_product_group, on=column, how="left")
  #Заполняем пропуски нулями
  merged_df = merged_df.fillna(0)
  #Создаем столбец Брак/Выпуск для определения процента брака
  merged_df['Брак,м2/Выпуск,м2'] = merged_df['Брак,м2']/merged_df['Выпуск,м2']*100
  #Считаем общий выпуск
  release_sum = merged_df['Выпуск,м2'].sum()
  #Добавляем столбец Выпуск,м2/Общий выпуск,м2
  merged_df['Выпуск,м2/Общий выпуск,м2'] = merged_df['Выпуск,м2']/release_sum*100
  #Считаем общий брак
  incongruity_sum = merged_df['Брак,м2'].sum()
  #Добавляем столбец Брак,м2/Общий брак,м2
  merged_df['Брак,м2/Общий брак,м2'] = merged_df['Брак,м2']/incongruity_sum*100
  #Сортируем датафрейм по убыванию колонки Выпуск
  merged_df = merged_df.sort_values('Выпуск,м2/Общий выпуск,м2', ascending=False)
  #Возвращаем результат
  return merged_df
```

```python
def pivot_table_for_collections(release_by_product,incongruity_by_product,type_products):
  #Создаем пустой датафрейм для таблицы пластиков и пленок
  pivot_plastic_films = pd.DataFrame()
  #Запускаем функцию для создания общей таблицы по пластикам и пленкам
  pivot_plastic_films = pivot_by_columns(release_by_product,incongruity_by_product,'Цвет пленки/пластика',type_products)
  #Создаем пустой датафрейм для таблицы изделий
  pivot_milling = pd.DataFrame()
  #Запускаем функцию для создания общей таблицы по изделиям
  pivot_milling = pivot_by_columns(release_by_product,incongruity_by_product,'Изделие',type_products)
  return pivot_plastic_films,pivot_milling
```

```python
#Функция создания словаря на основе двух списков
def proverka_data(list_proverka,list_data_brak):
  brk = {}
  for i in range(len(list_data_brak)):
    brk[list_data_brak[i]] = list_proverka[i]
  return brk
def proverka_data_frame_release_incongruity(df_release,df_incongruity):
  pivot_plastic_films = []
  pivot_milling = []
  for type_of_product in df_release['Тип_продукции'].unique():
    pivot_plastic_films.append(pivot_table_for_collections(df_release,df_incongruity,type_of_product)[0])
    pivot_milling.append(pivot_table_for_collections(df_release,df_incongruity,type_of_product)[1])
  #Список значений брака по коллекциям (сводные)
  sum_incongruity = []
  for type_of_product in df_release['Тип_продукции'].unique():
    sum_incongruity.append(proverka_data(pivot_plastic_films,df_release['Тип_продукции'].unique())[type_of_product]['Брак,м2'].sum())
  #Список значений брака по коллекциям (исходный датафрейм)
  sum_incongruity_proverka = []
  for type_of_product in df_release['Тип_продукции'].unique():
    sum_incongruity_proverka.append(df_incongruity[df_incongruity['Тип_продукции']==type_of_product]['Брак,м2'].sum())
  #Список значений выпуска по коллекциям (сводные)
  sum_release = []
  for type_of_product in df_release['Тип_продукции'].unique():
    sum_release.append(proverka_data(pivot_plastic_films,df_release['Тип_продукции'].unique())[type_of_product]['Выпуск,м2'].sum())
  #Список значений выпуска по коллекциям (исходный датафрейм)
  sum_release_proverka = []
  for type_of_product in df_release['Тип_продукции'].unique():
    sum_release_proverka.append(df_release[df_release['Тип_продукции']==type_of_product]['Выпуск,м2'].sum())
  #Словарь сумм брака по коллекциям (сводные)
  incongruity_pr_svod = proverka_data(sum_incongruity,df_release['Тип_продукции'].unique())
  #Словарь сумм брака по коллекциям (исходный)
  incongruity_pr_ish = proverka_data(sum_incongruity_proverka,df_release['Тип_продукции'].unique())
  #Словарь сумм выпуска по коллекциям (сводные)
  release_pr_svod = proverka_data(sum_release,df_release['Тип_продукции'].unique())
  #Словарь сумм выпуска по коллекциям (исходный)
  release_pr_ish = proverka_data(sum_release_proverka,df_release['Тип_продукции'].unique())
  return {"Брак сводные": incongruity_pr_svod,"Брак общая": incongruity_pr_ish,"Выпуск сводные": release_pr_svod,"Выпуск общая":release_pr_ish}
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Определение текущей даты и вычисление периода с 30 сентября 2023 по 30 сентября 2024
current_date = datetime.now()
start_date = datetime(2023, 9, 30)  # 30 сентября 2023
end_date = datetime(2024, 10, 1)     # 30 сентября 2024

# Форматирование дат
start_date_str = start_date.strftime('%Y-%m-%d')
end_date_str = end_date.strftime('%Y-%m-%d')

# Фильтрация данных по "Тип_продукции", "Месяц" и исключение Леруа Мерлен
filtered_release = vertical_concat[(vertical_concat['Контрагент'] != "ЛЕРУА МЕРЛЕН (ЛЕ МОНЛИД)") &
    (vertical_concat['Тип_продукции'] == 'Original') &
    (vertical_concat['Месяц'] >= start_date_str) &
    (vertical_concat['Месяц'] <= end_date_str)
]
# Фильтрация данных по "Тип_продукции", "Месяц" и исключение Леруа Мерлен
filtered_release_top3 = vertical_concat[
    (vertical_concat['Тип_продукции'] == 'Original') &
    (vertical_concat['Месяц'] >= start_date_str) &
    (vertical_concat['Месяц'] <= end_date_str)
]
filtered_brak = vertical_concat[(vertical_concat['Контрагент'] != "ЛЕРУА МЕРЛЕН (ЛЕ МОНЛИД)") &
    (vertical_concat['Тип_продукции'] == 'Original') &
    (vertical_concat['Месяц'] >= start_date_str) &
    (vertical_concat['Месяц'] <= end_date_str)
]
filtered_brak_top3 = vertical_concat[
    (vertical_concat['Тип_продукции'] == 'Original') &
    (vertical_concat['Месяц'] >= start_date_str) &
    (vertical_concat['Месяц'] <= end_date_str)
]


# Рассчет суммы по столбцам "Выпуск,м2" и "Брак,м2" без фильтрации по контрагенту
sum_vypusk = filtered_release_top3['Выпуск,м2'].sum()
sum_brak = filtered_brak_top3['Брак,м2'].sum()

# Рассчет суммы по столбцам "Выпуск,м2" и "Брак,м2" без фильтрации по контрагенту
sum_vypusk_bez_lm = filtered_release['Выпуск,м2'].sum()
sum_brak_bez_lm = filtered_brak['Брак,м2'].sum()

# Расчет среднего процента брака
avg_brak_percent = (sum_brak_bez_lm / sum_vypusk_bez_lm) * 100
print(f"Средний процент брака по пленкам, исключая Леруа Мерлен составляет {avg_brak_percent:.2f}%")

# Создание сводной таблицы по цвету пленки/пластика для брака
pivot_table_brak = pd.pivot_table(filtered_brak,
                                  values=['Брак,м2'],
                                  index='Цвет пленки/пластика',
                                  aggfunc='sum',
                                  fill_value=0)

# Создание сводной таблицы по цвету пленки/пластика для выпуска
pivot_table_release = pd.pivot_table(filtered_release,
                                     values=['Выпуск,м2'],
                                     index='Цвет пленки/пластика',
                                     aggfunc='sum',
                                     fill_value=0)



# Создание сводной таблицы по цвету пленки/пластика для брака
pivot_table_brak_top3 = pd.pivot_table(filtered_brak_top3,
                                  values=['Брак,м2'],
                                  index='Цвет пленки/пластика',
                                  aggfunc='sum',
                                  fill_value=0)

# Создание сводной таблицы по цвету пленки/пластика для выпуска
pivot_table_release_top3 = pd.pivot_table(filtered_release_top3,
                                     values=['Выпуск,м2'],
                                     index='Цвет пленки/пластика',
                                     aggfunc='sum',
                                     fill_value=0)
# Объединение двух сводных таблиц
pivot_table = pivot_table_release.join(pivot_table_brak, on='Цвет пленки/пластика', how='inner', lsuffix='_выпуск', rsuffix='_брак')

# Добавление столбца "Брак/Выпуск,%"
pivot_table['Брак/Выпуск,%'] = (pivot_table['Брак,м2'] / pivot_table['Выпуск,м2']) * 100

# Поиск пленок с процентом брака выше среднего
higher_than_avg_brak = pivot_table[pivot_table['Брак/Выпуск,%'] > avg_brak_percent]

# Сортировка по полю "Брак,м2" для нахождения топ 3 по браку
top_3_brak = higher_than_avg_brak.nlargest(3, 'Брак,м2')
name_of_films = []
# Выводим Топ 3 пленок по браку
print("Топ 3 пленок по браку:")
for index, row in top_3_brak.iterrows():
    print(f"Пленка {index}, брак, м2: {row['Брак,м2']}, выпуск, м2: {row['Выпуск,м2']}, Брак/Выпуск: {row['Брак/Выпуск,%']:.2f}%")
    name_of_films.append(index)
# Объединение двух сводных таблиц топ 3 по выпуску
pivot_table_top3 = pivot_table_release_top3.join(pivot_table_brak_top3, on='Цвет пленки/пластика', how='inner', lsuffix='_выпуск', rsuffix='_брак')

# Добавление столбца "Брак/Выпуск,%"
pivot_table_top3['Брак/Выпуск,%'] = (pivot_table_top3['Брак,м2'] / pivot_table_top3['Выпуск,м2']) * 100
# Сортировка по полю "Выпуск,м2" для нахождения топ 3 по выпуску
top_3_vypusk = pivot_table_top3.nlargest(3, 'Выпуск,м2')

# Выводим Топ 3 пленок по выпуску
print("\nТоп 3 пленок по выпуску:")
for index, row in top_3_vypusk.iterrows():
    print(f"Пленка {index}, выпуск, м2: {row['Выпуск,м2']}, брак, м2: {row['Брак,м2']}, Брак/Выпуск: {row['Брак/Выпуск,%']:.2f}%")
    name_of_films.append(index)
# Вывод периода
print(f"\nПериод анализа: {start_date.strftime('%d %B %Y')} - {end_date.strftime('%d %B %Y')}")

# Общий выпуск и брак за период по типу продукции
print(f"\nОбщий выпуск за период: {sum_vypusk:.2f} м2")
print(f"Общий брак за период: {sum_brak:.2f} м2")
```

```python
def get_defects_by_films(film_list, vertical_concat_type_of_brak, start_date_str, end_date_str):
    # Фильтрация данных по типу продукции и периоду
    filtered_brak = vertical_concat_type_of_brak[
        (vertical_concat_type_of_brak['Тип_продукции'] == 'Original') &
        (vertical_concat_type_of_brak['Месяц'] >= start_date_str) &
        (vertical_concat_type_of_brak['Месяц'] <= end_date_str) &
        (vertical_concat_type_of_brak['Цвет пленки/пластика'].isin(film_list))
    ]

    # Группировка по плёнке и виду брака
    defects_by_film = filtered_brak.groupby(['Цвет пленки/пластика', 'Вид брака'])['Брак,м2'].sum().reset_index()

    return defects_by_film

def get_products_by_films(film_list, vertical_concat_type_of_brak, vertical_concat_release, start_date_str, end_date_str):
    # Фильтрация данных по типу продукции и периоду
    filtered_brak = vertical_concat_type_of_brak[
        (vertical_concat_type_of_brak['Тип_продукции'] == 'Original') &
        (vertical_concat_type_of_brak['Месяц'] >= start_date_str) &
        (vertical_concat_type_of_brak['Месяц'] <= end_date_str) &
        (vertical_concat_type_of_brak['Цвет пленки/пластика'].isin(film_list))
    ]

    # Группировка по плёнке и изделию для брака
    products_by_film = filtered_brak.groupby(['Цвет пленки/пластика', 'Изделие'])['Брак,м2'].sum().reset_index()

    # Фильтрация данных по выпуску из другой таблицы
    filtered_release = vertical_concat_release[
        (vertical_concat_release['Тип_продукции'] == 'Original') &
        (vertical_concat_release['Месяц'] >= start_date_str) &
        (vertical_concat_release['Месяц'] <= end_date_str) &
        (vertical_concat_release['Цвет пленки/пластика'].isin(film_list))
    ]

    # Группировка по плёнке и изделию для выпуска
    release_by_film = filtered_release.groupby(['Цвет пленки/пластика', 'Изделие'])['Выпуск,м2'].sum().reset_index()

    # Объединение таблиц по изделиям и добавление столбца брак/выпуск
    merged_data = pd.merge(products_by_film, release_by_film, on=['Цвет пленки/пластика', 'Изделие'], how='left')
    merged_data['Брак/Выпуск,%'] = (merged_data['Брак,м2'] / merged_data['Выпуск,м2']) * 100

    return merged_data
```

```python
# Пример списка плёнок
film_list = name_of_films

# Определение периода для фильтрации
start_date_str = '2023-09-30'
end_date_str = '2024-09-30'

# Получение данных по видам брака
defects = get_defects_by_films(film_list, vertical_concat_type_of_brak, start_date_str, end_date_str)
print("Результаты по видам брака:")

for film, group in defects.groupby('Цвет пленки/пластика'):
    print(f"\nПленка {film}:")
    for index, row in group.iterrows():
        print(f"- Вид брака: {row['Вид брака']}, Площадь брака: {row['Брак,м2']} м2")

# Получение данных по изделиям, выпуску и соотношению брак/выпуск
products_with_release = get_products_by_films(film_list, vertical_concat_type_of_brak, vertical_concat_release, start_date_str, end_date_str)
print("\nРезультаты по изделиям:")

for film, group in products_with_release.groupby('Цвет пленки/пластика'):
    print(f"\nПленка {film}:")
    for index, row in group.iterrows():
        print(f"- Изделие: {row['Изделие']}, Выпуск: {row['Выпуск,м2']} м2, Брак: {row['Брак,м2']} м2, Брак/Выпуск: {row['Брак/Выпуск,%']:.2f}%")
```
